1) Crear una lista de asignaturas con los textos:matamticas, fisica, historia. Crear una lista de notas: 9,9,8 Crea una series con la lista notas como valores, y la lista asignaturas como indices.
```python
# Creamos las listas asignatura
asignaturas=['matematicas','fisica','historia']
# Creamos las listas notas
notas=['9','9','8']
#Creamos la serie donde las notas son los valores
serie_notas=pd.Series(notas,index=asignaturas)
#Creacion de la variable
nota_de_fisica = serie_notas["fisica"]

# Mostrar el mensaje
print(f"La nota de Física es {nota_de_fisica}")
```

3) Utilizando el DF de la F1 generado por el proceso de scraping, generar una gráfica con la distribución de las puntuaciones por paises. Nota: Utilice las librerias aquí utilizadas y asegurese de incluir una formato más elaborado que el presentado en la clase, es decir con más elementos de visualización.

```python
 #Usamos la función read_html  para leer el contenido de la pagina ESPN, donde se muestran las posiciones de F1 para la temporada 2023.
dataframe_f1 = pd.read_html("https://www.espn.com.mx/deporte-motor/f1/posiciones/_/temporada/2023")
# Mostramos la primera tabla
dataframe_f1[0]
```

```python
# Mostramos la segunda tabla
dataframe_f1[1]
```

```python
# Verificamos la estructura y los tipos de datos de la segunda tabla.
dataframe_f1[1].info()
```

```python
#Asignamos el dataframe de la segunda tabla a la variable df_paises_f1.
df_paises_f1 = dataframe_f1[1]
```

```python
#Mostramos las columnas del dataframe
df_paises_f1.columns
```

```python
# Remplazamos "-" del dataframe por un 0
for i in df_paises_f1.columns:
    df_paises_f1[i] = df_paises_f1[i].replace("-", 0)
```

```python
#Mostramos el dataframe sin los "-"
df_paises_f1
```

```python
# Cambiamos los valores nulos por un 0
df_paises_f1.fillna(0, inplace=True)
df_paises_f1
```

```python
# Definimos los tipos de datos de Dataframe como enteros
for i in df_paises_f1.columns:
    df_paises_f1[i] = df_paises_f1[i].astype(int)
```

```python
#Usamos el metodo loc para agregar una nueva fila al dataframe.
# La nueva fila contiene la suma de los valores numericos de todas las columnas
df_paises_f1.loc["Total"] = df_paises_f1.sum()
df_paises_f1
```

```python
# Seleccionamos la fila Total del dataframe y la almacenamos en una nueva variable.
# Esto nos da una Serie con la suma total.
df_pts_paises = df_paises_f1.loc["Total"]
#Convertimos la Serie df_pts_paises en un datadrame.
df = pd.DataFrame(df_pts_paises)
# Restablecemos el indice del dataframe.
# Esto convierte los indices  en una columna normal llamada index.
df.reset_index(inplace=True)
#Cambiamos el nombre de las columnas
df.columns = ["Pais", "Puntos"]
#Mostramos el dataframe
df
```

```python
# Librerias para graficas
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# plt.figure(figsize=(10, 6))
# sns.histplot(df["Puntos"], kde = True)
# plt.title("Distribución de Puntos por País")
# plt.xlabel("Puntos")
# plt.ylabel("Frecuencia")
# plt.show()
```

```python
fig, axs = plt.subplots(figsize=(15, 6))
axs.bar(x = df["Pais"], height = df["Puntos"])
axs.set_title("Puntos por País")
axs.set_xlabel("País")
axs.set_ylabel("Puntos")
```

  #### **Gráficos más elaborados**: 👩🏻‍💻🎨

```python
# Creamos el gráfico con tamaño (18,6)
plt.figure(figsize=(18, 6))
#Lo configuramos como un grafico de barras horizontales (barh)
#color='lightskyblue' es el color de las barras
#edgecolor='blue' es el color del contorno
plt.barh(df['Pais'], df['Puntos'], color='lightskyblue',edgecolor='blue')
# Agregamos etiquetas 'Pais' en el eje x con tamaño 20 y color naranja
plt.xlabel('Pais',fontsize = 20,color = 'orange')
# Agregamos etiquetas 'Puntos' en el ejer y con tamaño 20 y color verde
plt.ylabel('Puntos',fontsize = 20,color = 'green')
#Agregamos el titulo con tamaño 25 y colo cafe
plt.title('Distrbución por paises',fontsize = 25,color = 'brown')
# Mostramos el gráfico
plt.show()
```

```python
# Creamos el gráfico con tamaño (18,6)
plt.figure(figsize=(18, 6))
#Creamos con gráfico con la paqueteria seaborn (sns.barplot)
#palette="icefire" paleta que colores para el gráfico
#edgecolor='blue' edgecolor='blue'
grafica=sns.barplot(x=df['Pais'], y=df['Puntos'], palette="icefire",edgecolor='blue',width=0.75)
#Creamos la linea puntedad de color negro que sigue los valores del gráfico
sns.lineplot(x=df['Pais'], y=df['Puntos'], color='black',linewidth=2,linestyle='dashed')
# Agregamos etiquetas 'Pais' en el eje x con tamaño 20 y color dorado
plt.xlabel('Pais',fontsize = 20, color = 'darkgoldenrod')
# Agregamos etiquetas 'Pais' en el eje y con tamaño 20 y color verde
plt.ylabel('Valor',fontsize = 20, color = 'seagreen')
#Agregamos el titulo con tamaño 25 y colo azul
plt.title('Distrbucion por paises',fontsize = 25,color = 'blue')
#Ponemos etiquetas para escribir los puntos exactos de cada pais en el gráfico
for container in grafica.containers:
    grafica.bar_label(container, fontsize=12, color='crimson')
#Agregamos la interpretacion de la grafica
plt.figtext(0.1, 0, "Podemos observar que la mayoria de paises obtuvieron 102 puntos, Italia 0 pero existe un valor Italia2 con 102 puntos, Qatar 135, finalmente Azerbaiyán, Austria, Bélgica, Estados Unidos y Brasil 138 puntos"  , fontsize = 10)
# Mostramos el gráfico
plt.show()
```

```python
#Ordenamos los valores en forma ascendente
df1=df.sort_values('Puntos')
# Creamos el gráfico con tamaño (18,6)
# facecolor='thistle' color violeta pastel de fondo
plt.figure(figsize=(18, 6),facecolor='thistle')
#Creamos con gráfico con la paqueteria seaborn (sns.barplot)
#palette="Greens" paleta que colores para el gráfico
#edgecolor='blue' color de contorno
#width=0.95 ancho de las barras
grafica=sns.barplot(x=df1['Pais'], y=df1['Puntos'], palette="Greens",edgecolor='red',width=0.95)
#Creamos la linea puntedad de color negro que sigue los valores del gráfico
sns.lineplot(x=df1['Pais'], y=df1['Puntos'], color='black',linewidth=2,linestyle='dashed')
# facecolor='thistle' color violeta pastel de fondo
grafica.set_facecolor('thistle')
# Marcador de cruz para ITA con color rojo
grafica.plot('ITA', 3, "x", markersize=15, color="red")
# Marcador de triangulo para BRN con color verde
grafica.plot('BRN', 105, "^", markersize=20, color="green")
# Marcador de estrella para QAT con color plata
grafica.plot('QAT', 135, "*", markersize=20, color="silver")
# Marcador de estrella para BEL con color dorado
grafica.plot('BEL', 140, "*", markersize=20, color="darkgoldenrod")
# Agregamos etiquetas 'Pais' en el eje x con tamaño 20 y color azul
plt.xlabel('Pais',fontsize = 20, color = 'blue')
# Agregamos etiquetas 'Valor' en el eje y con tamaño 20 y color negro
plt.ylabel('Valor',fontsize = 20, color = 'black')
#Agregamos el titulo con tamaño 25 y colo azul
plt.title('Distrbución por paises ordenada de manera ascendente con marcadores en los cambios de valor',fontsize = 20,color = 'black')
# Mostramos el gráfico
plt.show()
```
