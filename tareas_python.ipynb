El objetivo sera, crear un dataframe con las ventas de nuestras 3 oficinas para los meses de enero, febrero y marzo.
Pasos a seguir:
Crea una variable "ventas_enero" que contenga una lista de 3 numeros enteros aleatorios entre 100 y 500
Crea una variable "ventas_febrero" que contenga una lista de 3 numeros enteros aleatorios entre 100 y 500
Crea una variable "ventas_marzo" que contenga una lista de 3 numeros enteros aleatorios entre 100 y 500
Crea una variable "filas" que contenga esta lista de palabras ["ventas_enero","ventas_febrero","ventas_marzo"]
Crea una variable "columnas" que contenga esta lista de palabras : "oficina1","oficina2","oficina3"
Crea una variable "ventas" que contenga una lista con los valores de ventas de enero, febrero y ventas_marzo
Crea una variable "dataframe" que contenga las ventas, filas y columnas

```python
ventas_enero=np.random.randint(100,501,3)
ventas_febrero=np.random.randint(100,501,3)
ventas_marzo=np.random.randint(100,501,3)
filas=["ventas_enero","ventas_febrero","ventas_marzo"]
columnas=["oficina1","oficina2","oficina3"]
ventas=[ventas_enero,ventas_febrero,ventas_marzo]
dataframe=pd.DataFrame(data=ventas, index=filas, columns=columnas)
dataframe
```

## Ejercicio 3

Utilizando el DF de la F1 generado por el proceso de scraping, generar una gráfica con la distribución de las puntuaciones por paises. Nota: Utilice las librerias aquí utilizadas y asegurese de incluir una formato más elaborado que el presentado en la clase, es decir con más elementos de visualización.

```python
f1=pd.read_html('https://www.espn.com.mx/deporte-motor/f1/posiciones/_/temporada/2023')
f1
```

```python
#Reemplazamos el valor de '-' por 0
df_f1=f1[1].replace('-'	, 0)
df_f1
```

```python
df_f1.info()
```

```python
columnas=df_f1.columns
columnas
```

```python
def transform_to_numeric(df):
  for i in columnas:
    df[i]=pd.to_numeric(df[i],errors='coerce')
  return df
```

```python
df_f1_limpio=transform_to_numeric(df_f1)
df_f1_limpio
```

```python
df_f1_limpio.info()
```

```python
df_f1_limpio.fillna(value=0, inplace=True)

df_f1_limpio
```

```python
df_f1_limpio.loc['Suma']=df_f1_limpio.sum()
df_f1_limpio
```

```python
df_country=pd.DataFrame(df_f1_limpio.loc['Suma'])
df_country
```

```python
df_country_columnas =df_country.reset_index()
df_country_columnas
```

### Grafica con Mahplolib

```python
import matplotlib.pyplot as plt
```

```python
# Crear el histograma
plt.figure(figsize=(18, 6))
plt.bar(df_country_columnas['index'], df_country_columnas['Suma'], edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Pais')
plt.ylabel('Valor')
plt.title('Distrbucion por paises')

# Mostrar el gráfico
plt.show()
```
